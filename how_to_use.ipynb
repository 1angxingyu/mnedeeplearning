```python
import mne_soft as sft
```

```python
sublist = []
for i in range(1,31):
    sublist += [f'{i}']#列表存储要分析的epoch序号，epoch需要命名为 subN-epo.fif
ROIS = {'F_ROI':['E6','E13','E112','E29','E111','E7','E30'],
        'C_ROI':['E105','E106','E31','E37','E80','E87','E54','E55','E79','E53'],
        'P_ROI':['E86','E60','E61','E78','E85','E62','E67','E77','E72']} #####用字典的方式存储感兴趣区以及它们包含的电极
conditions = ['CCC', 'CCU', 'CUC', 'CUU', 'CLC', 'CLU','NCO','NLO','NUN'] #划分epoch时的event_id，也就是对应的各个实验条件
#sfreq取决于采样率，这里也可以先降采样之后再设置
#n_per_seg是每个时间窗的大小，如果sfreq=512，也就是每秒512个点，n_per_seg=256表示时间窗为0.5s,n_overlap表示每个窗的重叠点数，可以根据步长计算：noverlap=nperseg-step
#n_per_seg最好是sfreq×2**N
#
```

```python
for sub in sublist:
    sft.get_amp_epoch(sub,ROIS, conditions,t_min=0.15,t_max=0.35,name='P200',csv_file_path='get_amp_epoch.csv')#每个ROI里的每个电极的每个条件每个epoch的P200
    sft.get_amp_epoch(sub,ROIS, conditions,t_min=0.35,t_max=0.45,name='N400',csv_file_path='get_amp_epoch.csv')#N400
    sft.get_amp_epoch(sub,ROIS, conditions,t_min=0.55,t_max=0.65,name='P600',csv_file_path='get_amp_epoch.csv')#P600
    sft.get_psd_epoch(sub, ROIS, conditions, 4, 7, csv_file_path='get_amp_epoch.csv', sfreq=1000, n_per_seg=500, n_overlap=250)#每个epoch里Theta频段功率
    sft.get_psd_epoch(sub, ROIS, conditions, 8, 12, csv_file_path='get_amp_epoch.csv', sfreq=1000, n_per_seg=500, n_overlap=250)#Alpha频段功率
    sft.get_psd_epoch(sub, ROIS, conditions, 12.5, 16, csv_file_path='get_amp_epoch.csv', sfreq=1000, n_per_seg=500, n_overlap=250)#lowBeta频段功率
    sft.get_psd_epoch(sub, ROIS, conditions, 16.5, 30, csv_file_path='get_amp_epoch.csv', sfreq=1000, n_per_seg=500, n_overlap=250)#hignBeta频段功率
#如果需要的是epoch里各个窗里的hsd而不是平均之后的整个epoch的hsd，就使用get_psd_window()
#能够得到26*7列数据，如果使用get_psd_window()分窗的话会有更多
```
